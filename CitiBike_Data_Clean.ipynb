```python
# Dependencies
import pandas as pd
import numpy as np
import glob
```

```python
#Read files and create df from "Happy Planet Index 2016" (Countries Grouped by region)
Bike_path= "./Data/2019_May_Jul_citibikenyc.xlsx"
Bike_read = pd.DataFrame(pd.read_excel(Bike_path, sheet_name='Sheet1'))

CitiBike_df = pd.DataFrame(Bike_read)
```

```python
CitiBike_df.dtypes
```

```python
CitiBike_df_grp=CitiBike_df.groupby(["Start_Date"])[["tripduration_Min"]].sum()
CitiBike_df_grp
```

```python
#Read files and create df from "Happy Planet Index 2016" (Countries Grouped by region)
Accident_path= "./Data/2019_May_Jul_BikeAccidentsNY.xlsx"
Accident_read = pd.DataFrame(pd.read_excel(Accident_path, sheet_name='Sheet1'))

Accident_df = pd.DataFrame(Accident_read)

Accident_df_grp=Accident_df.groupby(["CRASH DATE"])[["NUMBER OF CYCLIST INJURED"]].sum()
Accident_df_grp
```
